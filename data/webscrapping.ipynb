```python
import requests
from bs4 import BeautifulSoup

# Send a GET request to the website
url = 'https://www.td.com/ca/en/personal-banking'
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.text, 'html.parser')

# Extract relevant information (example: all paragraph text)
paragraphs = soup.find_all('p')
for paragraph in paragraphs:
    text = paragraph.get_text(strip=True)
    print(text)

# You would add additional code here to clean and preprocess the text,
# then save it in your desired format for model training.
```
